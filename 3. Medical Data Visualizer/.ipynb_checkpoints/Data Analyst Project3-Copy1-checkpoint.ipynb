# Heat Map

```python
df_heat=df.copy()
df_heat
```

```python
df_heat=df.loc[(df['ap_lo'] <= df['ap_hi'])&(df['height'] >= df['height'].quantile(0.025))&(df['height'] <= df['height'].quantile(0.975))]
df_heat
```

```python
cond1=(df['ap_lo'] <= df['ap_hi'])
cond2=(df['height'] >= df['height'].quantile(0.025))
cond3=(df['height'] <= df['height'].quantile(0.975))
cond4=(df['weight']>=df['weight'].quantile(0.025))
cond5=(df['weight']<=df['weight'].quantile(0.975))
```

```python
df_heat=df.loc[cond1&cond2&cond3&cond4&cond5]
df_heat.columns=['id','age','height','gender','height','ap_hi', 'ap_lo','cholesterol','gluc','smoke','alco','active','cardio','overweight']
```

```python
corr=df_heat.corr()
corr
```

```python
mask = np.triu(corr)
```

```python
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(corr, linewidths=1, annot=True, square=True, mask=mask, fmt=".1f",  center=0.08,cbar_kws = {"shrink":0.5}) 
#cbar_kws = {"shrink":0.5} )
```
